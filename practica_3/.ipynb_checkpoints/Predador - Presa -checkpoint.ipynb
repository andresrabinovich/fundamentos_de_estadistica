seguimos  usando  hoy.   El  modelo  de  Volterra  para  depredacion  comienza  suponiendo  la

tiene una dinamica `malthusiana', es decir, exponencial:  si no hay depredador, la poblacion

5)  Implemente la funcion
misVecinos(centro)
que tenga por input un
centro
,  corre-
spondiente a las coordenadas de un punto del tablero, y devuelva una matriz de 8

2
que tenga en sus  las las coordenadas vecinas al centro, en el orden determinado por
el siguiente esquema:

```r
misVecinos <- function(centro){
    vecinos <- matrix(0, ncol=2, nrow=8)
    vecinos[1, 1] <- centro[1] - 1
    vecinos[1, 2] <- centro[2] - 1
    vecinos[2, 1] <- centro[1] - 1
    vecinos[2, 2] <- centro[2]
    vecinos[3, 1] <- centro[1] - 1
    vecinos[3, 2] <- centro[2] + 1
    vecinos[4, 1] <- centro[1]
    vecinos[4, 2] <- centro[2] + 1
    vecinos[5, 1] <- centro[1] + 1
    vecinos[5, 2] <- centro[2] + 1
    vecinos[6, 1] <- centro[1] + 1
    vecinos[6, 2] <- centro[2]
    vecinos[7, 1] <- centro[1] + 1
    vecinos[7, 2] <- centro[2] - 1
    vecinos[8, 1] <- centro[1]
    vecinos[8, 2] <- centro[2] - 1    
    return(vecinos)
}
misVecinos(c(2, 2))
```

6) Implemente  la  funcion
buscarAdyacente(tablero, centro, esp)
que  toma  como
inputs un tablero (

), las coordenadas del centro de busqueda (

), 1 (

) o un 0 (lugar vaco).  La funcion debe devolver
las coordenadas del

vecino donde se encuentra la especie pedida, o el valor
"NO"
si la especie buscada no se encuentra en la vecindad.
En adelante, cada vez que necesitemos recorrer todo el tablero lo haremos
por  las, empezando desde el extremo izquierdo superior (
(2

2)
)

```r
buscarAdyacente <- function(tablero, centro, esp){
    vecinos <- misVecinos(centro)
    especie_pedida <- which(tablero[vecinos] == esp)
    if(length(especie_pedida) > 0){
        return(vecinos[especie_pedida[1], ])
    }else{
        return("NO")
    }
}
miTablero
str(buscarAdyacente(miTablero, c(2,5), -1))

```

7)  Implemente la funcion
indMueve(tablero)
, que tiene por input un tablero (
tablero
)
y ejecuta la fase de movimiento, recorriendo cada casillero del mismo.  En cada paso,
veri ca si la posicion (
i; j
) elegida esta ocupada por un bicho; en caso de estar ocupada,
buscamos el primer vecino desocupado y alli deplazamos al bicho en cuestion.

```r
indMueve <- function(tablero){
    for(i in 2:(nrow(tablero)-1)){
        for(j in 2:(ncol(tablero)-1)){
            if(tablero[i, j] %in% c(-1, 1)){
                nueva_posicion <- buscarAdyacente(tablero, c(i,j), 0)
                if(nueva_posicion[1] != "NO"){
                    #Marco la nueva posición de los bichos con un 2*tipo_de_bicho para no volver a moverlos una vez movidos
                    tablero[nueva_posicion[1], nueva_posicion[2]] <- 2*tablero[i, j]
                    tablero[i, j] <- 0
                }
            }
        }
    }
    tablero[tablero == 2] <- 1
    tablero[tablero == -2] <- -1
    return(tablero)
}
miTablero
indMueve(miTablero)
```

8)  Implemente la funcion
depredCome(tablero)
que tiene por input un tablero (
tablero
)
y ejecuta la fase de alimentacion recorriendo cada casillero del mismo.  En cada paso,
veri ca si la posicion (
i; j
) elegida esta ocupada por un tiburon y, en tal caso, busca la
primer foca dispoible en la vecindad.  Si la encuentra, el tiburon se mueve a la posicion
ocupada por la foca, y la devora (la foca desaparece del tablero).

```r
depredCome <- function(tablero){
    for(i in 2:(nrow(tablero)-1)){
        for(j in 2:(ncol(tablero)-1)){
            if(tablero[i, j] %in% c(-1)){
                nueva_posicion <- buscarAdyacente(tablero, c(i,j), 1)
                if(nueva_posicion[1] != "NO"){
                    #Marco la nueva posición de los tiburones con un -2 para no volver a moverlos una vez movidos
                    tablero[nueva_posicion[1], nueva_posicion[2]] <- -2
                    tablero[i, j] <- 0
                }
            }
        }
    }
    tablero[tablero == -2] <- -1
    return(tablero)
}
miTablero
depredCome(miTablero)
```

9)  Implemente la funcion
reproduccion(tablero)
que tiene por input un tablero (
tablero
)
y ejecuta la fase de reproduccion recorriendo cada casillero del mismo.  En cada paso,
veri ca si la posicion (
i; j
) elegida esta ocupada.  En caso de estarlo, busca en la vecin-
dad si hay otro individuo de su misma especie, en caso de encontrarlo, busca la primer
posicion vaca y hace aparecer en esa posicion un nuevo individuo de la misma especie
(foca o tiburon segun corresponda)

```r
reproduccion <- function(tablero){
    for(i in 2:(nrow(tablero)-1)){
        for(j in 2:(ncol(tablero)-1)){
            if(tablero[i, j] %in% c(-1, 1)){
                pareja <- buscarAdyacente(tablero, c(i,j), tablero[i, j])
                if(pareja[1] != "NO"){
                    #Generamos un nuevo bicho en esa posición
                    posicion_vacia <- buscarAdyacente(tablero, c(i,j), 0)
                    if(posicion_vacia[1] != "NO"){
                        tablero[posicion_vacia[1], posicion_vacia[2]] <- 2*tablero[i, j]
                    }else{
                        posicion_vacia <- buscarAdyacente(tablero, pareja, 0)
                        if(posicion_vacia[1] != "NO"){
                            tablero[posicion_vacia[1], posicion_vacia[2]] <- 2*tablero[i, j]
                        }
                    }
                    tablero[i, j] <- tablero[pareja[1], pareja[2]] <- 2*tablero[i, j]
                }
            }
        }
    }
    tablero[tablero == -2] <- -1
    tablero[tablero == 2]  <- 1
    return(tablero)
}
miTablero
reproduccion(miTablero)
```

10)  Simule la evolucion del sistema durante 2 ciclos partiendo del siguiente tablero inicial,
de nido en el item 4.

```r
imprimir <- function(elemento, verbose = TRUE){
    if(verbose) print(elemento)
}
ciclo <- function(tablero, ciclos, verbose=TRUE){
    imprimir("tablero inicial", verbose)
    imprimir(miTablero, verbose)
    for(i in 1:ciclos){
        imprimir(paste("Ciclo", i), verbose)
        tablero<-indMueve(tablero)
        imprimir("movimiento", verbose)
        imprimir(tablero, verbose)
        tablero<-depredCome(tablero) 
        imprimir("depredador come", verbose)    
        imprimir(tablero, verbose)
        tablero<-reproduccion(tablero)
        imprimir("se reproducen", verbose)
        imprimir(tablero, verbose)
    }
    return(tablero)
}
```
